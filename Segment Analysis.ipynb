```python
df['yummy'] = [1 if x=='Yes' else 0 for x in df['yummy']]
df['convenient'] = [1 if x=='Yes' else 0 for x in df['convenient']]
df['spicy'] = [1 if x=='Yes' else 0 for x in df['spicy']]
df['fattening'] = [1 if x=='Yes' else 0 for x in df['fattening']]
df['greasy'] = [1 if x=='Yes' else 0 for x in df['greasy']]
df['fast'] = [1 if x=='Yes' else 0 for x in df['fast']]
df['fast'] = [1 if x=='Yes' else 0 for x in df['fast']]
df['cheap'] = [1 if x=='Yes' else 0 for x in df['cheap']]
df['tasty'] = [1 if x=='Yes' else 0 for x in df['tasty']]
df['expensive'] = [1 if x=='Yes' else 0 for x in df['expensive']]
df['healthy'] = [1 if x=='Yes' else 0 for x in df['healthy']]
df['disgusting'] = [1 if x=='Yes' else 0 for x in df['disgusting']]
```

```python
# Replacing VisitFrequency 

df['VisitFrequency'] = df['VisitFrequency'].replace(to_replace= 'Once a month', value = 3) 
df['VisitFrequency'] = df['VisitFrequency'].replace(to_replace= 'Every three months', value = 2) 
df['VisitFrequency'] = df['VisitFrequency'].replace(to_replace= 'Once a year', value = 1) 
df['VisitFrequency'] = df['VisitFrequency'].replace(to_replace= 'Once a week', value = 4) 
df['VisitFrequency'] = df['VisitFrequency'].replace(to_replace= 'Never', value = 0) 
df['VisitFrequency'] = df['VisitFrequency'].replace(to_replace= 'More than once a week', value = 5)
                                                   
```

```python
# Replacing Gender 

df['Gender'] = df['Gender'].replace(to_replace= 'Female', value = 0) 
df['Gender'] = df['Gender'].replace(to_replace= 'Male', value = 1)
```

```python
#Using for loop for iterations from 1 to 10.  
# elbow method to find the best number of clusters for this data
from sklearn.cluster import KMeans 
k = range(2,15) 
sse = []
for i in k:
  model_demo = KMeans(n_clusters=i,random_state = 0)
  model_demo.fit(x)                                     #we get the sse value 
  sse.append(model_demo.inertia_) 
plt.figure(figsize = (15,7))
plt.scatter(k,sse)
plt.plot(k,sse)
plt.show()
```
